# Imports

```python
%load_ext autoreload
%autoreload 2

import numpy as np
import tensorflow as tf

from dataloader.utils import get_count_classes, plot_classes, plot_from_one_class

SEED = 0
```

# Create Dataset

```python
ds_train, ds_val = tf.keras.utils.image_dataset_from_directory(
    directory="tiny-imagenet-200/train/",
    batch_size=128,
    image_size=(64, 64),
    shuffle=True,
    seed=SEED,
    validation_split=0.2,
    subset="both"
)
```

# Class distribution

```python
class_counts = get_count_classes(ds_train)

class_counts.plot.hist(title="Distribution of class counts");
```
